```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('data/european_transition_paths.csv')
data = pd.melt(data, id_vars=["club","unit_name"])
data = data.rename(columns={"unit_name": "Country", "club": "Club", "variable": "Year", "value": "Relative transition path"})
data["Year"] = data["Year"].str.replace("X", "").astype(int)
data = data[data["Club"] != "divergent"]

country_dict = {}
with open("data/countries_and_codes.csv", encoding="utf-8") as file:
    for line in file:
        line = line.split(",")
        country_dict[line[1].strip()] = line[0].strip()

def code_to_country(code):
    return country_dict[code]

data["Country"] = data["Country"].apply(code_to_country)

fig, ax = plt.subplots(3, 1, figsize=(15, 14), sharex=True, sharey=True)

for i, club in enumerate(data["Club"].unique().tolist()):
    data_club = data[data["Club"] == club]
    palette = sns.color_palette("Set2", len(data_club["Country"].unique()))
    sns.lineplot(data_club, x='Year', y='Relative transition path', hue="Country", ax=ax[i], palette=palette)
    ax[i].set_ylabel('')  # Clear the y-axis label
    ax[i].legend(loc='upper left')
plt.show()
fig.figure.savefig("fig_output/Fig_05.png", dpi=600)
```
