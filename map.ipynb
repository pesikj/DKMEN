```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

shapefile_path = 'ne_10m_admin_0_countries/ne_10m_admin_0_countries.shp'
world = gpd.read_file(shapefile_path)
colormap_name = 'Set2'

country_dict = {}

with open("data/countries_and_codes.csv", encoding="utf-8") as file:
    for line in file:
        line = line.split(",")
        country_dict[line[1].strip()] = line[0].strip()

def code_to_country(code):
    return country_dict[code]

clubs = [
    "LU, IE, DK, SE, NL, AT, BE, DE, MT",
    "IT, CY, ES, SI, CZ, EE, SK, LT, PL",
    "PT, EL, HR, HU, LV, RO, BG",
]
divergent = "FI, FR"

cmap = plt.get_cmap(colormap_name)
num_colors = len(clubs) + 1

divergent = [code_to_country(x) for x in divergent.split(", ")]

country_codes = [[code_to_country(x) for x in country_codes.split(", ")] for country_codes in clubs]
all_countries = divergent.copy()

fig, ax = plt.subplots(1, 1, figsize=(15, 10))

for i, item in enumerate(country_codes):
    world[world['NAME'].isin(item)].plot(ax=ax, color=cmap(i/num_colors), edgecolor='black')
    all_countries += item
world[world['NAME'].isin(divergent)].plot(ax=ax, color=cmap(num_colors/num_colors), edgecolor='black')

# Plot non-EU countries in white
world[~world['NAME'].isin(all_countries)].plot(ax=ax, color='white', edgecolor='black')

# Adjust the plot limits to focus on Europe more closely
ax.set_xlim([-25, 40])
ax.set_ylim([34, 71])

legend_elements = [Patch(facecolor=cmap(i / num_colors), edgecolor='black', label=f"Club {i+1}") for i in range(len(clubs))]
legend_elements += [Patch(facecolor=cmap(num_colors / num_colors), edgecolor='black', label=f"Divergent Countries")]
    
ax.legend(handles=legend_elements, loc='upper left')

plt.show()
ax.figure.savefig("fig_output/Fig_06.png", dpi=600)
```
